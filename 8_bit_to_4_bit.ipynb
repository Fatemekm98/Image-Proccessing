```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# Load the image in grayscale
img = cv2.imread('/content/I19.BMP', cv2.IMREAD_GRAYSCALE)

# Convert to 4-bit image
img_4bit = np.floor_divide(img, 16)


fig=plt.figure(dpi=200)

fig.add_subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.axis("off")
plt.title("8-bit")

fig.add_subplot(1,2,2)
plt.imshow(img_4bit,cmap='gray')
plt.axis("off")
plt.title("4-bit")

# Save the new image
cv2.imwrite('image_4bit.jpg', img_4bit)
```

```python
from skimage import data, img_as_float
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import structural_similarity as ms_ssim
from skimage.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import skimage.io


img = skimage.img_as_float(img)

mae_const = mean_absolute_error(img, img_4bit)
mse_const = mean_squared_error(img, img_4bit)
ssim_const = ssim(img, img_4bit,
                  data_range=img_4bit.max() - img_4bit.min())
ms_ssim_const = ms_ssim(img, img_4bit)


print('The MAE index between the original and converted images is:', mae_const)
print('The MSe index between the original and converted images is:', mse_const)
print('The SSIM index between the original and converted images is:', ssim_const)
print('The MS_SSIM index between the original and converted images is:', ms_ssim_const)
```

```python
# Load the image in grayscale
pic = cv2.imread('/content/I19.BMP', cv2.IMREAD_GRAYSCALE)

# Convert to 4-bit image by splitting 8-bit values into two 4-bit values
pic_4bit = ((pic & 0b1111) << 4) + (pic >> 4)


fig=plt.figure(dpi=200)

fig.add_subplot(1,2,1)
plt.imshow(pic,cmap='gray')
plt.axis("off")
plt.title("8-bit")

fig.add_subplot(1,2,2)
plt.imshow(pic_4bit,cmap='gray')
plt.axis("off")
plt.title("4-bit")

# Save the new image
cv2.imwrite('pic_4bit.jpg', pic_4bit)
```

```python
# Load the image in grayscale
fk = cv2.imread('/content/I19.BMP', cv2.IMREAD_GRAYSCALE)

# Convert to 4-bit image by splitting 8-bit values into two 4-bit values
height, width = fk.shape
fk_4bit = np.zeros((height, width), dtype=np.uint8)
for i in range(height):
    for j in range(width):
        fk_4bit[i,j] = (fk[i,j] // 16) * 16


fig=plt.figure(dpi=200)

fig.add_subplot(1,2,1)
plt.imshow(fk,cmap='gray')
plt.axis("off")
plt.title("8-bit")

fig.add_subplot(1,2,2)
plt.imshow(fk_4bit,cmap='gray')
plt.axis("off")
plt.title("4-bit")

# Save the new image
cv2.imwrite('fk_4bit.jpg', fk_4bit)
```
