```python
thresh1 = cv2.threshold(gray_pic1 , 20, 255, cv2.THRESH_BINARY)[1]
print(thresh1)
```

```python
w=b=0
for i in range(1280):
  for j in range(1006):
     p=thresh1[i][j]
     if p==255:
      w=w+1
     else:
       b=b+1

print('white pixels:', w)
print('black pixels:', b)
```
