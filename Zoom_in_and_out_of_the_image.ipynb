```python
import cv2,matplotlib
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
pic = cv2.imread('/content/sign green.jpg')
print (pic)
```

```python
pic.shape
```

```python
gray_pic = cv2.cvtColor(pic , cv2.COLOR_RGB2GRAY)
plt.imshow(cv2.cvtColor(gray_pic, cv2.COLOR_GRAY2RGB))
```

```python
gray_pic.shape
```

```python
print(gray_pic)
```

```python
thresh = cv2.threshold(gray_pic ,100, 1, cv2.THRESH_BINARY)[1]
print(thresh)
```

```python
import cv2 as cv
img1 = cv.resize(pic,None,fx=0.5, fy=0.5, interpolation = cv.INTER_NEAREST)
plt.imshow(img1)
```

```python
img2 = cv.resize(img1,None,fx=2, fy=2, interpolation = cv.INTER_NEAREST)
plt.imshow(img2)
```
